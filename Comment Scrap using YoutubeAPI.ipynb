```python
from apiclient.discovery import build
import json
# Arguments that need to passed to the build function
DEVELOPER_KEY = "AIzaSyCpXi54I"
YOUTUBE_API_SERVICE_NAME = "youtube"
YOUTUBE_API_VERSION = "v3"

# creating Youtube Resource Object
youtube_object = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION,developerKey = DEVELOPER_KEY)
print(youtube_object)

search_keyword = youtube_object.search().list(q = 'One Peice', part = "id, snippet",maxResults = 10).execute()

# extracting the results from search response
results = search_keyword.get("items", [])
print(json.dumps(results[0]))

```
