```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def Check_file(filename):
    filename = pd.read_csv(filename)
    categoricalcol = [x for x in filename.dtypes.index if filename.dtypes[x]=='object']
    numericatypecol = [x for x in filename.dtypes.index if filename.dtypes[x]=='int64']
    floattypecol =  [x for x in filename.dtypes.index if filename.dtypes[x]=='float64']
    
    def df_plot_categorical(filename):
        data = pd.DataFrame(filename, columns=categoricalcol)
        for col in data.columns:
            fig = plt.figure(figsize = (13, 5))
            sns.countplot(data[col])
            plt.show()
    print("Categorical Plots",df_plot_categorical(filename))        

    def df_plot_numerical(filename):
        data1 = pd.DataFrame(filename, columns=numericatypecol)
        for col in data1.columns:
            fig = plt.figure(figsize = (13, 5))
            sns.distplot(data1[col])
            plt.show()
    print("Categorical Plots",df_plot_numerical(filename))
    return filename.head()
    
```

```python
Check_file('heart.csv')
```
